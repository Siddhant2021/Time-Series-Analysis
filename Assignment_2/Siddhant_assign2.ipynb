```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
df=pd.read_csv("Assignment 2  - Q2.csv")
df.head()
```

```python
x=df['Hrsspent']
y=df['Marks']
plt.scatter(x,y)
m, b = np.polyfit(x, y, 1)
y_cap=m*x+b
plt.plot(x,y_cap,color='red')
print("Eq of line is β0+β1x where \n β1->",m,"& β0->",b)
```

```python
SSE=(y-y_cap)**2
SSE=SSE.sum()
y_mean=y.mean()
SST=(y-y_mean)**2
SST=SST.sum()
r2=(1-SSE/SST)
r2
```

```python
data=pd.read_csv('Assignment 2  - Q3.csv')
data.head()
```

```python
X=data['x']
Y=data['y']
plt.scatter(X,Y)
```

```python
n=100
x_bar=X.mean()
y_bar=Y.mean()
sumxy=X*Y
sumxy=sumxy.sum()
sumx=X.sum()
sumy=Y.sum()
sumx2=X*X
sumx2=sumx2.sum()
sumx_2=sumx**2
B1=(n*sumxy -sumx*sumy)/(n*sumx2-sumx_2)
B0=y_bar-B1*x_bar
print("Eq of line is β0+β1x where \n β1->",B0,"& β0->",B1)

```

```python
Y_cap=B0+B1*X
plt.scatter(X,Y)
plt.plot(X,Y_cap,color='red')
```

```python
RSS=(Y-Y_cap)**2
RSS=RSS.sum()
TSS=(Y-y_bar)**2
TSS=TSS.sum()
r2=(1-RSS/TSS)
r2
```

This is the Ecommerce Customers csv file from a company. It has Customer info, suchas Email, Address, and their color Avatar. Then it also has numerical value columns:

* Avg. Session Length: Average session of in-store style advice sessions.
* Time on App: Average time spent on App in minutes
* Time on Website: Average time spent on Website in minutes
* Length of Membership: How many years the customer has been a member. 

```python
customers=pd.read_csv('Ecommerce Customers')
customers.head()
```

```python
customers.describe()
```

```python
customers.info()
```

```python
import seaborn as sns
```

```python
sns.pairplot(customers,plot_kws={"color":"rosybrown","edgecolor":"black"},diag_kws={"color":"rosybrown","edgecolor":"black"})
```

So we can clearly see that Length of Membership and Yearly Amount Spent are the most correlated features

```python
sns.lmplot(x='Yearly Amount Spent',y='Length of Membership',data=customers,line_kws={'color': 'skyblue'},scatter_kws={'color': 'pink','edgecolor':'black'})
```

Training and Testing Data

```python
x=customers[['Avg. Session Length',	'Time on App',	'Time on Website','Length of Membership']]
y=customers[['Yearly Amount Spent']]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(x_train,y_train)
```

```python
predictions=lm.predict(x_test)
```

```python
plt.scatter(y_test,predictions,color='darkgrey',edgecolors='black',s=30)
```

```python
from sklearn.metrics import r2_score
r2=r2_score(y_test,predictions)
r2
```

```python
df2= pd.DataFrame( data = lm.coef_.reshape(4), columns = ['Coefficient'] ,index = x_train.columns)
df2
```

Interpreting the coefficients:

- Holding all other features fixed, a 1 unit increase in **Avg. Session Length** is associated with an **increase of 25.98 total dollars spent**.
- Holding all other features fixed, a 1 unit increase in **Time on App** is associated with an **increase of 38.59 total dollars spent**.
- Holding all other features fixed, a 1 unit increase in **Time on Website** is associated with an **increase of 0.19 total dollars spent**.
- Holding all other features fixed, a 1 unit increase in **Length of Membership** is associated with an **increase of 61.27 total dollars spent**.
